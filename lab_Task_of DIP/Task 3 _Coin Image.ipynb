#  Task: Display RGB Components, Grayscale, and Binary Image

###  Objective:
To read a colored image and:
- Display its **Red**, **Green**, and **Blue** components separately.
- Convert the image into **Grayscale**.
- Convert the Grayscale image into a **Binary** image.

###  Input Image:
**Filename:** `Coin.jpg`

###  Steps Involved:
1. Read the image using OpenCV.
2. Split it into RGB channels.
3. Convert the image into grayscale.
4. Apply thresholding to obtain the binary image.
5. Display all results using Matplotlib.


```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Step 1: Read the colored image
img = cv2.imread('Coin.jpg')
```

```python
# Step 2: Convert from BGR (OpenCV default) to RGB for correct color display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Step 3: Split the image into Red, Green, and Blue components
R, G, B = cv2.split(img_rgb)
```

```python
# Step 4: Convert the image to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Step 5: Convert Grayscale to Binary (Thresholding)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
# Step 6: Display all results
plt.figure(figsize=(15,8))

plt.subplot(2,3,1)
plt.title("Original RGB Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(2,3,2)
plt.title("Red Component")
plt.imshow(R, cmap='Reds')
plt.axis('off')

plt.subplot(2,3,3)
plt.title("Green Component")
plt.imshow(G, cmap='Greens')
plt.axis('off')

plt.subplot(2,3,4)
plt.title("Blue Component")
plt.imshow(B, cmap='Blues')
plt.axis('off')

plt.subplot(2,3,5)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(2,3,6)
plt.title("Binary Image")
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
